**Condicional If**

```python
# Condicional If
if 5 > 2:
    print("A senteça é verdadeira!")
```

```python
# Condicional If...Else
if 5 < 2:
    print("A sentença é verdadeira!")
else:
    print("A sentença é falsa!")
```

```python
# Condicional If...Else com variável
dia = "Terça"
if dia == "Segunda":
    print("Hoje fará sol!")
else:
    print("Hoje vai chover!")
```

```python
# Podemos usar o operdor elif para validar mais de uma condição
if dia == "Segunda":
    print("Hoje fará sol!")
elif dia == "Terça":
    print("Hoje vai chover!")
else:
    print("Sem previsão do tempo para o dia selecionado")
```
