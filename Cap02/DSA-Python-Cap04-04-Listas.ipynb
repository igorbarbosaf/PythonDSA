**Operador in**

```python
#Criando uma lista
lista_teste_op = 100, 2, -5, 3.4
```

```python
# Verificando se o valor 10 pertence a lista
print(10 in lista_teste_op)
```

```python
# Verificando se o valor 100 pertence a lista
print(100 in lista_teste_op)
```
