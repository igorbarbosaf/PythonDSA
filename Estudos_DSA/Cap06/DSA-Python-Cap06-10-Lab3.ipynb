**Expressões Regulares**

Expressões regulares são padrões usados para combinar ou encontrar ocorrências de sequências de caracteres em uma string. Em Python, expressões regulares são geralmente usadas para manipular strings e realizar tarefas com validação de entrada de dados, extração de informações de strings e substituição de texto.

```python
import re # Importa o módulo re para trabalhar com expressões regulares
```

```python
texto = "Meu e-mail é exemplo@gmail.com e você pode me contar em outro_email@yahoo.com."
```

```python
# Expressão regular para contar quantas vezes o caracter arroba aparece no texto
resultado = len(re.findall("@", texto))# findall() é uma função do módulo re que encontra todas as ocorrências de um padrão em uma string e retorna uma lista de todas as ocorrências encontradas.
```

```python
print("O caractere '@' apareceu", resultado, "vezes no texto.")
```

```python
# Expressão regular para extrair a palavra que aparece após a palavra "você" em um texto
resultado = re.findall(r'você (\w+)', texto)
```

```python
print("A palavra após 'você' é:", resultado[0])
```

Nota: O r antes da string que representa a expressão regular em Python é usado para indicar que a string é uma string literal raw. Isso significa que as barras invertidas() não interpretadas como caracteres de escape, mas são incluídas na expressão regular como parte do padrão

```python
# Expressão regular para extrair endereços de e-mail de uma string
emails = re.findall(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b', texto)
```

```python
print(emails)
```

```python
text = "O aluno estava incrivelmente perdido, mas encontrou a DSA e rapidamente começou a aprender."
```

```python
# Extrair os advérbios da frase
for m in re.finditer(r"\w+mente\b", text):# finditer() é uma função do módulo re que encontra todas as ocorrências de um padrão em uma string e retorna um iterador de todos os resultados encontrados.
    print('%02d-%02d: %s' % (m.start(), m.end(), m.group(0)))
```
