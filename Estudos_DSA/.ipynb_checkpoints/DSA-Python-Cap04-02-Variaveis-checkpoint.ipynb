**Variáveis Atribuídas a Outras Variáveis e Ordem dos Operadores**

```python
largura = 2
```

```python
altura = 4
```

```python
area = largura * altura
```

```python
area
```

```python
perimetro = 2 * largura + 2 * altura
```

```python
perimetro
```

```python
# A ordem dos operadores é a mesma seguida na Matemática
perimetro = 2 * (largura + 2) * altura
```

```python
perimetro
```
