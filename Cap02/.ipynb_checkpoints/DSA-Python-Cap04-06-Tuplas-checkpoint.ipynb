```python
# Criando um tupla
tupla1 = ("Geografia", 23, "Elefantes", 9.8, 'Python')
```

```python
# Imprimindo a tupla
tupla1
```

```python
# Tuplas não suportam append()
tupla1.append("Chocolate")
```

```python
# Tuplas não suportam delete de um item específico
del tupla1["Elefantes"]
```

```python
# Tuplas podem ter um único item
tupla1 = ("Chocolate")
```

```python
tupla1
```

```python
tupla1 = ("Geografia", 23, "Elefantes", 9.8, 'Python')
```

```python
tupla1[0]
```

```python
# Verificando o comprimento da tupla
len(tupla1)
```

```python
# Slicing, da mesma forma que se faz com listas
tupla1[1:]
```

```python
tupla1.index('Elefantes')
```

```python
# Tuplas não suportam atribuição de item
tupla1[1] = 21
```

```python
# Deletando a tupla
del tupla1
```

```python
tupla1
```

```python
# Criando uma tupla
t2 = ('A', 'b', 'c')
```

```python
t2
```

```python
# Tuplas não suportam atribuição de item
t2[0] = 'D'
```

```python
# Usando a função list() para converter uma tupla para lista
lista_t2 = list(t2)
```

```python
type(t2)
```

```python
type(lista_t2)
```

```python
lista_t2
```

```python
lista_t2.append('D')
```

```python
# Usando a funçã0 tuple() para converter uma lista para tupla
t2 = tuple(lista_t2)
```
