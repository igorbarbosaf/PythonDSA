**Indexando String**

```python
# Atribuindo uma string
s = 'Data Science Academy'
```

```python
print(s)
```

Indexação em Python começar por zero.

```python
# Primeiro elemento da string
s[0]
```

```python
s[1]
```

```python
s[2]
```

```python
s[3]
```

```python
s[4]
```

Podemos usar um : para executar um slicing que faz leitura de tudo até um ponto designado. Por exemplo:

```python
# Retorna todos os elementos da string, começando pela posição
# (lembre-se que Python começa a indexação pela posição 0),
# até o fim da string.
s[1:]
```

```python
# A string original permance inalterda
s
```

```python
# Retorna tudo até a posição 3
s[:3]
```

```python
# Retorna tudo até a posição 4
s[:4]
```

```python
# Nós também podemos usar a indexação negativa e ler de trás para frente
s[-1]
```

```python
# Retornar tudo, exceto a últuma letra
s[:-1]
```

Nós também podemos usar a notação de índice e fatiar a string em pedaços específios (o padrão é 1).

Por exemplo, podemos usar dois pontos duas vezes em uma linha e, em seguida, um número que especifica a frequência para retornar elemntos.

```python
s[::1]
```

```python
s[::2]
```

```python
s[::-1]
```
