Gere código Python que crie uma lista com os números entre 1 e 100 e então imprima os números pares, mas somente se o número for divisível por 4.

```python
# Cria uma lista com números de 1 a 100
numeros = list(range(1, 101))
```

```python
type(numeros)
```

```python
# Itera pela lista e imprime os números pares que são divisíveis por 4
for numero in numeros:
    if numero % 2 == 0 and numero % 4 == 0:
        print(numero)
```

Gere código Python que crie uma lista com os números entre 1 e 100 e então imprima os números pares, mas somente se o número for divisível por 4, usando list comprehension.

```python
# Cria a lista de números divisíveis por 4 e pares entre 1 e 100 usando list comprehension
numeros_divisiveis_por_4 = [numero for numero in range(1, 101) if numero % 2 == 0 and numero % 4 == 0]

# Imprime os números que satisfazem a condição
print(numeros_divisiveis_por_4)
```
